以下是 Python 中各種運算子的範例，包括算術運算子、賦值運算子、比較運算子、邏輯運算子、成員運算子、身分運算子和位元運算子：

| 類型         | 運算子範例 |
|:---:         |:---:       |
| **算術運算子** | `+`, `-`, `*`, `/`, `//`, `%`, `**`<br>範例: `a = 5 + 3`, `b = 10 / 2`, `c = 2 ** 3` |
| **賦值運算子** | `=`, `+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=`<br>範例: `x = 10`, `x += 5` |
| **比較運算子** | `==`, `!=`, `>`, `<`, `>=`, `<=`<br>範例: `if a > b:`, `if x == 10:` |
| **邏輯運算子** | `and`, `or`, `not`<br>範例: `if a > 5 and b < 10:`, `if not is_valid:` |
| **成員運算子** | `in`, `not in`<br>範例: `if 'apple' in fruits:`, `if x not in [1, 2, 3]:` |
| **身分運算子** | `is`, `is not`<br>範例: `if a is b:`, `if x is not None:` |
| **位元運算子** | `&`, `|`, `^`, `~`, `<<`, `>>`<br>範例: `a & b`, `c << 2` |

### 詳細範例

**算術運算子：**
```python
a = 5
b = 3
print(a + b)  # 加法: 8
print(a - b)  # 減法: 2
print(a * b)  # 乘法: 15
print(a / b)  # 除法: 1.6667
print(a // b) # 整數除法: 1
print(a % b)  # 取餘數: 2
print(a ** b) # 指數: 125
```

**賦值運算子：**
```python
x = 10
x += 5  # 等於 x = x + 5，即 x 現在是 15
x *= 2  # 等於 x = x * 2，即 x 現在是 30
```

**比較運算子：**
```python
a = 5
b = 10
if a < b:
    print("a 小於 b")
else:
    print("a 不小於 b")
```

**邏輯運算子：**
```python
a = True
b = False
if a and not b:
    print("a 為 True 且 b 為 False")
if a or b:
    print("a 或 b 至少有一個為 True")
```

**成員運算子：**
```python
fruits = ['apple', 'banana', 'cherry']
if 'apple' in fruits:
    print("水果列表中有蘋果")
if 'grape' not in fruits:
    print("水果列表中沒有葡萄")
```

**身分運算子：**
```python
a = [1, 2, 3]
b = a
if a is b:
    print("a 和 b 是同一個物件")
if a is not [1, 2, 3]:
    print("a 和另一個列表不是同一個物件")
```

**位元運算子：**
```python
a = 60  # 二進位表示為 0011 1100
b = 13  # 二進位表示為 0000 1101
print(a & b)  # 位元與: 12 (0000 1100)
print(a | b)  # 位元或: 61 (0011 1101)
print(a ^ b)  # 位元異或: 49 (0011 0001)
print(~a)     # 位元取反: -61 (1100 0011)
print(a << 2) # 位元左移: 240 (1111 0000)
print(a >> 2) # 位元右移: 15 (0000 1111)
```

這些範例展示了如何使用不同類型的運算子來進行各種運算。

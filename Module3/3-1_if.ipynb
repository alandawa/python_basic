在 Python 中，`if-elif-else` 結構用於根據條件執行不同的程式碼塊。以下是如何使用這些條件語句的範例：

### if 語句

`if` 語句用於執行一個程式碼塊，當條件為真時。

**語法**:
```python
if condition:
    # 當條件為真時執行這裡的程式碼
```

**範例**:
```python
age = 18
if age >= 18:
    print("成年人")
```

### if...else 語句

`if...else` 語句在條件為真時執行一個程式碼塊，條件為假時執行另一個程式碼塊。

**語法**:
```python
if condition:
    # 當條件為真時執行這裡的程式碼
else:
    # 當條件為假時執行這裡的程式碼
```

**範例**:
```python
age = 16
if age >= 18:
    print("成年人")
else:
    print("未成年人")
```

### if...elif...else 語句

`if...elif...else` 語句用於處理多個條件，其中 `elif` 可以用來檢查多個其他條件，而 `else` 用於處理所有其他情況。

**語法**:
```python
if condition1:
    # 當 condition1 為真時執行這裡的程式碼
elif condition2:
    # 當 condition1 為假且 condition2 為真時執行這裡的程式碼
else:
    # 當所有其他條件都為假時執行這裡的程式碼
```

**範例**:
```python
score = 85
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("D")
```

### 綜合範例

以下是綜合使用 `if`, `if...else`, 和 `if...elif...else` 的範例：

```python
temperature = 25

# if
if temperature > 30:
    print("很熱")

# if...else...
if temperature > 30:
    print("很熱")
else:
    print("不熱")

# if...elif...else...
if temperature > 30:
    print("很熱")
elif temperature > 20:
    print("溫暖")
else:
    print("寒冷")
```

### 小結

- **`if`**: 用於單一條件的判斷。
- **`if...else`**: 用於兩種情況的判斷，一種條件為真，一種條件為假。
- **`if...elif...else`**: 用於多條件的判斷，可以處理多個可能的情況。
